```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# Read the list of car brands from the CSV file
car_brands_df = pd.read_csv("C:\\Users\morch\OneDrive\Desktop\Finsire\car_models.csv")  # Change the path accordingly
car_brands = car_brands_df['Brands'].tolist()  # Assuming the CSV has a column named 'Brands'
```

```python
# Define your ChromeDriver
driver = webdriver.Chrome()

# Base URL
base_url = "https://www.cardekho.com/newcars/"
```

```python
# Initialize a list to store the extracted data
all_car_data = []

# Iterate over each car brand
for brand in car_brands:
    url = base_url + brand
    driver.get(url)

    # Wait for the page to fully load
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, "body")))

    # Extract car models and their prices
    cars = driver.find_elements(By.CSS_SELECTOR, "ul.modelList > li")

    for car in cars:
        try:
            # Extract car model name
            model_name = car.find_element(By.CSS_SELECTOR, "h3 a").text

            # Extract car price
            price = car.find_element(By.CSS_SELECTOR, "div.price").text

            # Add the extracted data to the list
            all_car_data.append({
                "brand": brand,
                "model": model_name,
                "price": price
            })
        except Exception as e:
            print(f"An error occurred while extracting data for a car in brand {brand}: {e}")

# Convert the list of dictionaries to a DataFrame
all_car_data_df = pd.DataFrame(all_car_data)
```

```python
# Save the data to a new CSV file
output_csv_file = "all_car_models.csv"
all_car_data_df.to_csv(output_csv_file, index=False)

# Print a confirmation message
print("Data has been saved to the CSV file.")
```
