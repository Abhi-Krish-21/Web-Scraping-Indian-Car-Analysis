```python
car_brands = df1["brand"].unique()

# Function to remove brand names from model names
def remove_brand_name(model_name, brands):
    for brand in brands:
        if model_name.startswith(brand):
            return model_name[len(brand):].strip()
    return model_name

# Apply the function to the "model" column
df1["model"] = df1.apply(lambda row: remove_brand_name(row["model"], car_brands), axis=1)

# Display the updated DataFrame
print(df1)
```
